```python
def split_tokenize_write(all_texts, filepath='all_clauses.pkl'):
    all_clauses = []
    num = 1
    for text in all_texts:
        print(str(num)+'/'+str(len(all_texts)))
        num += 1
        clauses = [tokenize(i) for i in clause_split(text) if i and i != '\n' and i != ' ']
        all_clauses.extend(clauses)
    with open(filepath, 'wb') as outfile:
        pickle.dump(all_clauses, outfile)      
```

```python
def collect_wikinews(links1, links2):
    all_texts = []
    for link in links1:
        article_links = collect_links1(link)
        num = 1
        for article_link in article_links:
            print(str(num)+'/'+str(len(article_links)))
            num += 1
            all_texts.append(text_from_wiki_page(article_link))
    print('-------------------------------------------------------------')
    for link in links2:
        article_links = collect_links2(link)
        num = 1
        for article_link in article_links:
            print(str(num)+'/'+str(len(article_links)))
            num += 1
            all_texts.append(text_from_wiki_page(article_link))
    split_tokenize_write(all_texts)
```

```python
def get_href_from_td(soup):
    links = []
    list_tags = soup.find_all('td')
    for tag in list_tags:
        try:
            links.append((tag.find('a').get('href')))
        except:
            pass
    return links
```

```python
def collect_links1(link):
    root = requests.get(link)
    html = root.text
    soup = BeautifulSoup(html, 'lxml')
#     needed = soup.find('tbody')
    links = get_href_from_td(soup)
    return links
```

```python
def get_href_from_li(soup):
    links = []
    list_tags = soup.find_all('li')
    for tag in list_tags:
         links.append((tag.find('a').get('href')))
    return links
```

```python
def collect_links2(link):
    root = requests.get(link)
    html = root.text
    soup = BeautifulSoup(html, 'lxml')
    needed = soup.find('td', {'style': 'vertical-align:top;width:100%'})
    links = get_href_from_li(needed)
    return links
```

```python
def text_from_wiki_page(link, url='https://ru.wikinews.org'):
    texts = []
    page = requests.get(url+link)
    html = page.text
    soup = BeautifulSoup(html, 'lxml')
    needed = soup.find('div', {'class':'mw-parser-output'})
    if needed:
        p_tags = needed.find_all('p')
        for tag in p_tags:
            p_text = tag.text
            texts.append(p_text)
    text = ' '.join(texts)
    text = re.sub('\[.+?\]', '', text)
    return text
```

```python
SPLIT_REG = re.compile('([!,;:\.?—\n]+)')
def read_gold(filename):
    with open(os.path.join('GOLD', filename), 'r') as file:
        lines = file.readlines()
    all_clauses = []
    current_clause = []
    for line in lines:
        fields = line.split('\t')
        if len(fields) > 2:
            token = fields[2].lower().strip()
            if not SPLIT_REG.match(token):
                current_clause.append(token)
                
            else:
                current_clause.append(token)
                all_clauses.append(current_clause)
                current_clause = []
    return all_clauses
```

```python
# добавим гикря
for filename in os.listdir('GOLD'):
    current_clauses = read_gold(filename)
    clauses.extend(current_clauses)
```

```python
with open('all_clauses.pkl', 'wb') as outfile:
        pickle.dump(clauses, outfile)      
```
