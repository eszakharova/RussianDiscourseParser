```python
import numpy as np
import pandas as pd
import pickle
import regex as re
from preprocessing.parse_rs3_make_objects import EDUPair, EDU
from make_features import load_pairs_target, rel_type_only
```

```python
from nltk import ngrams
```

```python
from collections import Counter
```

```python
pairs,target = load_pairs_target() 
```

```python
target_rel = rel_type_only(target)
```

```python
all_dict = {}
for rel_type in set(target_rel):
    all_tokens = []
    for i in range(len(pairs)):
        if target_rel[i] == rel_type:
            all_tokens.extend(pairs[i].edu1.lemmatized_tokens)
            all_tokens.extend(pairs[i].edu2.lemmatized_tokens)
    all_dict[rel_type] = Counter(all_tokens)
for rel in all_dict:
    print(rel)
    print('\n'.join(map(str, all_dict[rel].most_common(50))))
    print('----------------')
```

```python
all_dict = {}
for rel_type in set(target_rel):
    all_bigrams = []
    for i in range(len(pairs)):
        if target_rel[i] == rel_type:
            all_bigrams.extend(ngrams(pairs[i].edu1.lemmatized_tokens, 2))
            all_bigrams.extend(ngrams(pairs[i].edu2.lemmatized_tokens, 2))
    all_dict[rel_type] = Counter(all_bigrams)
```

```python
for rel in all_dict:
    print(rel)
    print('\n'.join(map(str, all_dict[rel].most_common(50))))
    print('----------------')
```
